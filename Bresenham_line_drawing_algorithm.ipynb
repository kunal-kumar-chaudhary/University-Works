```python
x0 = int(input("enter the starting x cordinate"))
x1 = int(input("enter the final x cordinate"))
y0 = int(input("enter the starting y cordinate"))
y1 = int(input("enter the final y cordinate"))

# calculating dx and dy
dx = x1-x0
dy = y1-y0

# declaring two empty list so that we can store our suceessive coordinates in it.
x_cor = []
y_cor = []

# calculating the decision parameter
pk = 2 * dy - dx

# appending the initial points into the list
x_cor.append(x0)
y_cor.append(y0)

# applying the conditionals
for i in range(dx):
  if pk<0:
    pk_final = pk + 2 * dy
    x_next = x0 + 1
    y_next = y0
    x_cor.append(x_next)
    y_cor.append(y_next)
    x0 = x_next
    y0 = y_next

  else:
    pk_final = pk + (2*dy - 2*dx) 
    x_next = x0 + 1
    y_next = y0 + 1
    x_cor.append(x_next)
    y_cor.append(y_next)
    x0 = x_next
    y0 = y_next


print(x_cor) # printing the list containing x coordinates
print(y_cor) # printing the list containing y coordinates
```

```python
# now plotting the graph using matplotlib 
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,8))
plt.plot(x_cor,y_cor)
plt.title('line after deriving the point using line drawing algorithm')
plt.xlabel('x coordinate')
plt.ylabel('y coordinate')
```
